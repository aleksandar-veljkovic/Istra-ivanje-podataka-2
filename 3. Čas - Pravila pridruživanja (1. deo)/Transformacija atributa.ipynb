```python
import pandas as pd
```

```python
data = pd.read_csv('responses.csv')
```

```python
data
```

```python
data.isna().sum().values
```

```python
data = data.fillna(value=0)
```

```python
data.isna().sum().sum()
```

```python
data = data.replace({ 1: 'Low', 2: 'Low', 3: 'Medium', 4: 'High', 5: 'High' })
columns = data.columns
transactions = []
for index, row in data.iterrows():
    itemlist = []
    for col in columns:
        if col not in ['Height', 'Weight', 'Age', 'Number of siblings']:
            value = row[col]
            if value != 0:
                itemlist.append(f'{col}={value}')
            
    transactions.append(itemlist)
```

```python
from apyori import apriori
```

```python
res = apriori(transactions, min_support=0.1, min_confidence=0.5, min_lift=3)
for record in res:
#     print(record.items)
    if len(record.items) > 1:
        base = list(record.ordered_statistics[0].items_base)
        if len(base) > 0:
            items = list(record.ordered_statistics[0].items_add)
            support = round(record.support * 100, 2)
            confidence = round(record.ordered_statistics[0].confidence * 100, 2)
            lift = round(record.ordered_statistics[0].lift, 2)

            print(f'{", ".join(base)} => {", ".join(items)}[S={support}%, C={confidence}%, L={lift}]')
```
